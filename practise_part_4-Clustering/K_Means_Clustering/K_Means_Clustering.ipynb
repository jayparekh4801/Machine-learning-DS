# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Importing Dataset

```python
dataset = pd.read_csv("./Mall_Customers.csv")
independent = dataset.iloc[ : , [3, 4]].values
```

# Use Elbow Method To Find Optimal Clusters

```python
from sklearn.cluster import KMeans
wcss = []
for i in range(1, 11) :
    kmeans = KMeans(n_clusters = i, init = "k-means++", random_state = 42)
    kmeans.fit(independent)
    wcss.append(kmeans.inertia_)

plt.plot(range(1, 11), wcss)
plt.title("The Elbow Method")
plt.xlabel("Number Of Clusters")
plt.ylabel("WCSS")
plt.show()
```

# Traininig The KMeans Model On Dataset

```python
kmeans = KMeans(n_clusters = 5, init = "k-means++", random_state = 42)
dependent_kmeans = kmeans.fit_predict(independent)
```

```python
print(dependent_kmeans)
```

# Visualising The Clusters

```python
plt.scatter(independent[dependent_kmeans == 0, 0], independent[dependent_kmeans == 0, 1], s=100, c="red", label = "Cluster 1")
plt.scatter(independent[dependent_kmeans == 1, 0], independent[dependent_kmeans == 1, 1], s=100, c="blue", label = "Cluster 2")
plt.scatter(independent[dependent_kmeans == 2, 0], independent[dependent_kmeans == 2, 1], s=100, c="green", label = "Cluster 3")
plt.scatter(independent[dependent_kmeans == 3, 0], independent[dependent_kmeans == 3, 1], s=100, c="cyan", label = "Cluster 4")
plt.scatter(independent[dependent_kmeans == 4, 0], independent[dependent_kmeans == 4, 1], s=100, c="magenta", label = "Cluster 5")
plt.scatter(kmeans.cluster_centers_[ : , 0], kmeans.cluster_centers_[ : , 1], s = 100, c = "black", label = "centroids")
plt.title("The Clusters Of Customers")
plt.xlabel("Annual Income (k$)")
plt.ylabel("Spending Score (1-100)")
plt.legend()
plt.show()
```
