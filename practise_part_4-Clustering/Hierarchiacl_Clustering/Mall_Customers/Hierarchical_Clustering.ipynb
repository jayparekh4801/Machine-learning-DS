# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Importing Dataset

```python
dataset = pd.read_csv("./Mall_Customers.csv")
independent = dataset.iloc[ : , [3, 4]].values
```

# Using The Denderogram To Find Optimal Number Of Clusters

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(independent, method="ward"))
plt.title("Dendrogram")
plt.xlabel("Customers")
plt.ylabel("Euclidean Distance")
plt.show()
```

# Training Hierarchical Clustering Model On Dataset 

```python
from sklearn.cluster import AgglomerativeClustering
hc = AgglomerativeClustering(n_clusters = 5, affinity = "euclidean", linkage = "ward")
dependent_hc = hc.fit_predict(independent)
```

```python
print(dependent_hc)
```

# Vsualising Hierarchical Clustering Model

```python
plt.scatter(independent[dependent_hc == 0, 0], independent[dependent_hc == 0, 1], s = 100, c = "red", label = "Cluster 1")
plt.scatter(independent[dependent_hc == 1, 0], independent[dependent_hc == 1, 1], s = 100, c = "blue", label = "Cluster 2")
plt.scatter(independent[dependent_hc == 2, 0], independent[dependent_hc == 2, 1], s = 100, c = "cyan", label = "Cluster 3")
plt.scatter(independent[dependent_hc == 3, 0], independent[dependent_hc == 3, 1], s = 100, c = "yellow", label = "Cluster 4")
plt.scatter(independent[dependent_hc == 4, 0], independent[dependent_hc == 4, 1], s = 100, c = "black", label = "Cluster 5")
plt.title("clusters of customers")
plt.xlabel("Annual Income")
plt.ylabel("Spending Score")
plt.legend()
plt.show()
```
