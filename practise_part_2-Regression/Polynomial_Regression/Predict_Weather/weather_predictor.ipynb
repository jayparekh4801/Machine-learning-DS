# Improting Libraries

```python
import numpy as np
import pandas as pd
```

# Importing Dataset

```python
dataset = pd.read_excel("./AirQualityUCI.xlsx")
```

```python
import datetime as dt
dataset["Date"] = pd.to_datetime(dataset["Date"])
dataset["Date"] = dataset["Date"].map(dt.datetime.toordinal)
dataset = dataset.drop(["Time"], axis = 1)
```

```python
dataset
```

```python
independent = dataset.iloc[ : , : 11].values
dependent = dataset.iloc[ : , 11 : 14].values
```

```python
independent
```

```python
dependent
```

# Split Dataset Into Training Set And Test Set

```python
from sklearn.model_selection import train_test_split
independent_train, independent_test, dependent_train, dependent_test = train_test_split(independent, dependent, test_size = 0.2, random_state = 0)
```

# Training Polynomial Regression Model From Training Set

```python
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
regressor_poly = LinearRegression()
poly_fea = PolynomialFeatures(degree = 3)
independent_train_poly = poly_fea.fit_transform(independent_train)
independent_test_poly = poly_fea.fit_transform(independent_test)
regressor_poly.fit(independent_train_poly, dependent_train)
```

# Predicting Future Values

```python
dependent_predict = regressor_poly.predict(independent_test_poly)
```

```python
dependent_predict
```

```python
dependent_test
```

# Evaluating Performance Of Model

```python
from sklearn.metrics import r2_score
r2_score(dependent_test, dependent_predict)
```
