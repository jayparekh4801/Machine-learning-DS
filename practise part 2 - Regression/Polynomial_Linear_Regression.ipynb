# Importing Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Importing Datasets

```python
dataset = pd.read_csv("/Users/jayparekh/Documents/machine_learning_dataScience/Machine Learning A-Z (Codes and Datasets)/Part 2 - Regression/Section 6 - Polynomial Regression/Python/Position_Salaries.csv")
independent = dataset.iloc[ : , 1 : -1].values
dependent = dataset.iloc[ : , -1].values
```

# Making Simple Linear Regression Model from DataSet

```python
from sklearn.linear_model import LinearRegression
regressor_simple = LinearRegression()
regressor_simple.fit(independent, dependent)
```

# Making Polynomial Linear Regression Model From DataSet

```python
from sklearn.preprocessing import PolynomialFeatures
poly_fea = PolynomialFeatures(degree = 6)
independent_poly = poly_fea.fit_transform(independent)
regressor_poly = LinearRegression()
regressor_poly.fit(independent_poly, dependent)
```

# Visualising Linear Regeression Model

```python
plt.scatter(independent, dependent, color = "red")
plt.plot(independent, regressor_simple.predict(independent), color = "blue")
plt.title("Truth Or Bluff(Simple Linear Regression)")
plt.xlabel("levels")
plt.ylabel("salaries")
plt.show()
```

```python
plt.scatter(independent, dependent, color = "red")
plt.plot(independent, regressor_poly.predict(independent_poly), color = "blue")
plt.title("Truth Or Bluff(Polynomial Linear Regression)")
plt.xlabel("levels")
plt.ylabel("salaries")
plt.show()
```

```python
regressor_simple.predict([[6.5]])
```

# Visualising Polynomial Regression Curve With Higher Resolution

```python
indepnedent_grid = np.arange(min(independent), max(independent), 0.1)
indepnedent_grid = indepnedent_grid.reshape(len(indepnedent_grid), 1)
plt.scatter(independent, dependent, color = "red")
plt.plot(indepnedent_grid, regressor_poly.predict(poly_fea.fit_transform(indepnedent_grid)), color = "blue")
plt.title("Truth Or Bluff(Polynomial Linear Regression)")
plt.xlabel("levels")
plt.ylabel("salaries")
plt.show()
```

# Predicting A New Result With Linear Regression  

```python
regressor_simple.predict([[6.5]])
```

# Predictiong A New Result With Polynomial Regression 

```python
single_value = np.array(6.5, ndmin = 2)
regressor_poly.predict(poly_fea.fit_transform(single_value))
```

# Evaluating Performance Of Model

```python
from sklearn.metrics import r2_score
r2_score(dependent, dependent)
```
